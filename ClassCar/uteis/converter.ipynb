```python
from PIL import Image
from pathlib import Path

def png_para_ico(png_path, out_path=None, tamanhos=(16,24,32,48,64,128,256),
                 ajustar_quadrado=True, branco_transparente=False):
    """
    Converte um PNG em .ICO multi-tamanhos.
    - ajustar_quadrado: centraliza em um canvas quadrado transparente se o PNG n√£o for quadrado.
    - branco_transparente: tenta tornar branco puro (#FFFFFF) transparente (opcional).
    """
    png_path = Path(png_path)
    if out_path is None:
        out_path = png_path.with_suffix(".ico")

    img = Image.open(png_path).convert("RGBA")

    if branco_transparente:
        # Torna branco puro transparente (sem precisar instalar libs extras)
        datas = img.getdata()
        novas = []
        for r,g,b,a in datas:
            if r == 255 and g == 255 and b == 255:
                novas.append((255,255,255,0))
            else:
                novas.append((r,g,b,a))
        img.putdata(novas)

    if ajustar_quadrado:
        w, h = img.size
        if w != h:
            lado = max(w, h)
            canvas = Image.new("RGBA", (lado, lado), (0, 0, 0, 0))
            canvas.paste(img, ((lado - w)//2, (lado - h)//2), img)
            img = canvas

    tamanhos = sorted(set(int(s) for s in tamanhos))
    pares = [(s, s) for s in tamanhos]
    img.save(out_path, format="ICO", sizes=pares)
    return str(out_path)
```

```python
from pathlib import Path
for p in Path(".").glob("*.png"):
    print("Gerado:", png_para_ico(p))
```
